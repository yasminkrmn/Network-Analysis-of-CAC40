```python
tickers = pd.read_html('https://www.dividendmax.com/market-index-constituents/cac-40')[0]
tickers
```

```python
tickers = tickers.Ticker.to_list()
```

```python
tickers = [ticker + ".PA" for ticker in tickers]
No_hold = ["AIR.PA", "ATO.PA", "GLE.PA"]
tickers = [ticker for ticker in tickers if ticker not in No_hold]
tickers[0:5]
```

```python
frames = []

for ticker in tickers: 
    var = yf.Ticker(ticker)
    frame = var.institutional_holders
    #print(frame, var) 
    frame["comp"] = var.ticker
    frames.append(frame)
    
```

```python
frames
```

```python
all_stock = pd.concat(frames)
all_stock.head()
```

```python
G = nx.from_pandas_edgelist(all_stock, 'Holder', 'comp', edge_attr=True)
```

```python
edgelist = nx.to_edgelist(G)
edgelist
```

```python
dict(G.degree())
```

```python
colors = []
for node in G:
    if node in all_stock["comp"].values:
        colors.append("red")
    else:
        colors.append("blue")
```

```python
f = plt.figure(figsize=(50,50))
nx.draw(G, with_labels=True,
       node_color=colors,
       node_size=[v * 100 for v in dict(G.degree()).values()],
       width=[v[2]['Value'] / 300000000 for v in edgelist])
ax = plt.gca()
ax.collections[0].set_edgecolor("#696969")
f.savefig("network_analysis_of_CAC40.png")
```
